```python
# Cargar librerías necesarias
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset de ejemplo 'tips' de Seaborn
df = sns.load_dataset('tips')

# Mostrar las primeras filas del dataset para confirmación
df.head()
```

```python
# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(12, 6))

# Gráfico de dispersión con cuadrícula, leyenda, anotaciones y flechas
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='smoker', s=100, ax=ax)
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)

# Ajuste de la leyenda fuera del gráfico
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize='large', shadow=True, title='Detalles')

# Añadir cuadrícula
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')

# Añadir anotación para el mayor tip en 'Dinner'
ax.annotate('Mayor tip en la cena', xy=(50, 10), xytext=(45, 12),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Añadir flecha verde
ax.arrow(35, 7, 10, 1, head_width=0.8, head_length=0.5, fc='green', ec='green')

# Mostrar el gráfico
plt.show()
```

```python
# Cargar el dataset
df = sns.load_dataset('tips')

# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(12, 6))

# Definir una paleta de colores personalizada
custom_palette = {'Lunch': 'red', 'Dinner': 'blue'}  

# Gráfico de dispersión con la paleta personalizada, leyenda y anotaciones
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='smoker', s=100, ax=ax, palette=custom_palette)
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)

# Ajuste de la leyenda fuera del gráfico
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize='large', shadow=True, title='Detalles')

# Añadir cuadrícula
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')

# Añadir anotación para el mayor tip en 'Dinner'
ax.annotate('Mayor tip en Dinner', xy=(50, 10), xytext=(45, 12),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Añadir flecha verde
ax.arrow(35, 7, 10, 1, head_width=0.8, head_length=0.5, fc='green', ec='green')

# Mostrar el gráfico
plt.show()
```
