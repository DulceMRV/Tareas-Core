```python
import pandas as pd
import matplotlib.pyplot as plt
import io

# Datos de ejemplo en formato CSV
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""

# Cargar los datos en un DataFrame de Pandas
df = pd.read_csv(io.StringIO(data))
```

```python
# Convertir la columna 'fecha' a formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])
```

```python
# Grfico de lneas de las ventas
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['ventas'], marker='o')  # Agregar marcador para mejor visualizacin
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejorar la legibilidad
plt.tight_layout()
plt.show()
```

```python
# Crear un grfico de dispersin
plt.figure(figsize=(10, 6))
plt.scatter(df['fecha'], df['ventas'], color='orange')  # Cambiar el color para diferenciar del grfico de lneas
plt.title("Ventas Diarias (Dispersin)")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejor visualizacin
plt.tight_layout()
plt.show()
```

```python
# Crear un histograma de las ventas
plt.figure(figsize=(10, 6))
plt.hist(df['ventas'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribucin de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()
```
