```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Ruta del archivo
ruta_archivo = 'C:/Users/dulce/Downloads/retail_sales_dataset.csv'

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(ruta_archivo)

# Ver las primeras filas del DataFrame para confirmar la carga
print(df.head())
```

ESTADÍSTICAS DESCRIPTIVAS

```python
# Estadísticas descriptivas
print(df.describe())
```

HISTOGRAMA

```python
# Histograma para visualizar la distribución de las variables numéricas
plt.figure(figsize=(10, 6))
df['Age'].hist(bins=10, color='purple', edgecolor='black')
plt.title("Distribución de Edad")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Histograma para visualizar la distribución del Precio por Unidad
plt.figure(figsize=(10, 6))
df['Price per Unit'].hist(bins=10, color='orange', edgecolor='black')
plt.title("Distribución de Precio por Unidad")
plt.xlabel("Precio por Unidad")
plt.ylabel("Frecuencia")
plt.show()
```

BOXPLOT

```python
# Boxplot para detectar valores atípicos en el Total Amount
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='Total Amount', color='purple')
plt.title("Boxplot de Monto Total")
plt.ylabel("Monto Total")
plt.show()
```

LÍNEA DE TENDENDENCIA DE VENTA A LO LARGO DEL TIEMPO

```python
# Línea de tendencias de ventas a lo largo del tiempo
plt.figure(figsize=(12, 6))
df.groupby('Date')['Total Amount'].sum().plot()
plt.title("Tendencias de Ventas a lo Largo del Tiempo")
plt.xlabel("Fecha")
plt.ylabel("Monto Total de Ventas")
plt.show()
```

DISPERSIÓN: PRECIO POR UNIDAD Y CANTIDAD

```python
# Gráfico de dispersión entre Precio por Unidad y Cantidad
plt.figure(figsize=(10, 6))
plt.scatter(df['Price per Unit'], df['Quantity'], color='purple', alpha=0.6)
plt.title("Relación entre Precio por Unidad y Cantidad")
plt.xlabel("Precio por Unidad")
plt.ylabel("Cantidad")
plt.show()
```

```python
# Histogramas y boxplots combinados
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.histplot(df['Total Amount'], bins=15, kde=True, color='green', ax=axes[0])
axes[0].set_title("Histograma de Monto Total")

sns.boxplot(data=df, y='Total Amount', color='orange', ax=axes[1])
axes[1].set_title("Boxplot de Monto Total")
plt.show()
```
