CARGAR DATOS DE KAGGLE

```python
import pandas as pd
import os

# Ruta base de la carpeta "World Happiness Report"
base_path = "C:/Users/Dulce/Downloads/World Happiness Record/"

# Cargar cada archivo en un DataFrame y guardarlos en un diccionario
dataframes = {}
for year in ["2015", "2016", "2017", "2018", "2019"]:
    file_path = os.path.join(base_path, f"{year}.csv")
    dataframes[year] = pd.read_csv(file_path)

# Exploración inicial para verificar la estructura de los datos
for year, df in dataframes.items():
    print(f"\nAño {year}:")
    print(df.info())
    print(df.describe())
```

```python
# Diccionario de nombres de variables por año
variables_dict = {
    "2015": ["Happiness Score", "Economy (GDP per Capita)", "Health (Life Expectancy)", "Freedom"],
    "2016": ["Happiness Score", "Economy (GDP per Capita)", "Health (Life Expectancy)", "Freedom"],
    "2017": ["Happiness.Score", "Economy..GDP.per.Capita.", "Health..Life.Expectancy.", "Freedom"],
    "2018": ["Score", "GDP per capita", "Healthy life expectancy", "Freedom to make life choices"],
    "2019": ["Score", "GDP per capita", "Healthy life expectancy", "Freedom to make life choices"]
}
```

HISTOGRAMA

```python
# Generar histogramas con caritas felices para cada año y cada variable
for year, df in dataframes.items():
    print(f"\nHistogramas con caritas felices para el año {year}")
    variables = variables_dict[year]
    for variable in variables:
        plt.figure(figsize=(8, 6))
        
        # Crear histograma
        n, bins, patches = plt.hist(df[variable].dropna(), bins=20, color='yellow', edgecolor='black', alpha=0.7)
        
        # Agregar caritas felices sobre cada barra
        for count, x in zip(n, bins):
            plt.text(x + (bins[1] - bins[0]) / 2, count, "😊" * int(count // 5), ha='center', va='bottom', fontsize=10)
        
        # Títulos y etiquetas
        plt.title(f"Histograma de {variable} - {year}")
        plt.xlabel(variable)
        plt.ylabel("Frequency")
        
        plt.show()
```

```python
# Crear histogramas y boxplots en una sola figura para cada variable y cada año
for year, df in dataframes.items():
    print(f"\nAnálisis combinado para el año {year}")
    variables = variables_dict[year]
    for variable in variables:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
        
        # Histograma
        ax1.hist(df[variable].dropna(), bins=20, color='pink', edgecolor='black', alpha=0.7)
        ax1.set_title(f"Histogram of {variable} - {year}")
        ax1.set_xlabel(variable)
        ax1.set_ylabel("Frequency")
        
        # Boxplot
        ax2.boxplot(df[variable].dropna(), patch_artist=True, boxprops=dict(facecolor="cyan"))
        ax2.set_title(f"Boxplot of {variable} - {year}")
        ax2.set_xlabel(variable)
        
        plt.suptitle(f"Análisis de {variable} para el año {year}")
        plt.show()
```
